```python
import pandas as pd
```

```python
effects_100 = pd.read_csv("/FastData/czirion/Crypto_Diversity_Pipeline/analyses/variants/data/effects/HIGH_VNI_VNII_VNBI_VNBII_no-poly_80_no-mut_20_0.5_1.5_0.2.csv", header = 0, low_memory = False)
effects_80 = pd.read_csv("/FastData/czirion/Crypto_Diversity_Pipeline/analyses/variants/data/effects/HIGH_VNI_VNII_VNBI_VNBII_no-poly_80_no-mut_20_0.5_1.5_0.2.csv", header = 0, low_memory = False)
```

```python
lofs = pd.read_csv("/FastData/czirion/Crypto_Diversity_Pipeline/Crypto_Desjardins/results_2024-10-22/02.Dataset/snps/lofs.tsv", sep = "\t", header = 0, low_memory = False)
```

```python
lof_effects = lofs.merge(effects_100, on = ["var_id", "lineage"], how = "inner")
```

```python
lof_effects
```

```python
effects_100['var_id_surviving'] = effects_100['var_id']
effects_100.loc[effects_100['status'] != 'Variant', 'var_id_surviving'] = float('nan')
gene_strain_num_vars_100 = effects_100.groupby(["lineage", "strain", "gene_id", "gene_name", "description", "effect_type"])['var_id_surviving'].nunique().reset_index(name='num_vars')
gene_strain_num_vars_100['presence_vars'] = gene_strain_num_vars_100['num_vars'].apply(lambda x: 0 if x == 0 else 1)
strains_per_gene_100 = gene_strain_num_vars_100.groupby(["lineage", "gene_id", 'gene_name', 'description', 'effect_type'])['presence_vars'].sum().reset_index(name='num_strains')
strains_per_gene_100.sort_values(by='num_strains', ascending=False, inplace=True)
strains_per_gene_100
```

```python
effects_80['var_id_surviving'] = effects_80['var_id']
effects_80.loc[effects_80['status'] != 'Variant', 'var_id_surviving'] = float('nan')
gene_strain_num_vars_80 = effects_80.groupby(["lineage", "strain", "gene_id", "gene_name", "description", "effect_type"])['var_id_surviving'].nunique().reset_index(name='num_vars')
gene_strain_num_vars_80['presence_vars'] = gene_strain_num_vars_80['num_vars'].apply(lambda x: 0 if x == 0 else 1)
strains_per_gene_80 = gene_strain_num_vars_80.groupby(["lineage", "gene_id", 'gene_name', 'description', 'effect_type'])['presence_vars'].sum().reset_index(name='num_strains')
strains_per_gene_80.sort_values(by='num_strains', ascending=False, inplace=True)
strains_per_gene_80
```

```python
difference = strains_per_gene_100.merge(strains_per_gene_80, on = ["lineage", "gene_id", 'gene_name', 'description', 'effect_type'], how = "outer", suffixes = ('_100', '_80'))
```
